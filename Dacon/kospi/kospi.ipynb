```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

plt.rc('axes', unicode_minus=False)
```

```python
kospi = pd.read_csv('kospi_data.csv')
kospi.head()
```

```python
kospi.info()
```

```python
kospi['Year'] = pd.to_datetime(kospi['Date']).dt.year
kospi.head()
```

```python
kospi['Month'] = pd.to_datetime(kospi['Date']).dt.month
kospi['Day'] = pd.to_datetime(kospi['Date']).dt.day
kospi.head()
```

```python
for col in kospi.columns[1:7]:
    plt.figure(figsize=(4,4))
    sns.displot(kospi[col])
    plt.title(col)
    plt.show()
```

```python
sns.heatmap(kospi.corr(), annot=True)
plt.show()
```
