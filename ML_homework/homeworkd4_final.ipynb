```python
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import tensorflow as tf
import warnings
warnings.filterwarnings('ignore')
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
display(train.head(3))
display(test.head(3))
```

```python
train.isnull().sum()
```

```python
mean = train['Age'].mean()
print(mean)
```

```python
train['Age'] = train['Age'].fillna(mean)
train['Embarked'] = train['Embarked'].fillna(method='ffill')
train_df = train.drop(['Cabin'], axis=1)

train_df.isnull().sum()test_df.isnull().sum()
```

```python
test.isnull().sum()
```

```python
test['Age'] = test['Age'].fillna(test['Age'].mean())
test['Fare'] = test['Fare'].fillna(test['Fare']).mean()
test_df = test.drop(['Cabin'], axis=1)
```

```python
test_df.isnull().sum()
```

```python
encoder = LabelEncoder()
train_df['Embarked'] = encoder.fit_transform(train_df['Embarked'])
train_df['Sex'] = encoder.fit_transform(train_df['Sex'])
test_df['Embarked'] = encoder.fit_transform(test_df['Embarked'])
test_df['Sex'] = encoder.fit_transform(test_df['Sex'])
```

```python
print(train_df.info(), test_df.info())
```

```python
train_df.loc[train_df['Age']<8, 'Age'] = 0
train_df.loc[(train_df['Age'] >= 8) & (train_df['Age'] < 20), 'Age'] = 1
train_df.loc[(train_df['Age'] >= 20) & (train_df['Age'] < 60), 'Age'] = 2
train_df.loc[train_df['Age'] >= 60, 'Age'] = 3

test_df.loc[test_df['Age']<8, 'Age'] = 0
test_df.loc[(test_df['Age'] >= 8) & (test_df['Age'] < 20), 'Age'] = 1
test_df.loc[(test_df['Age'] >= 20) & (test_df['Age'] < 60), 'Age'] = 2
test_df.loc[test_df['Age'] >= 60, 'Age'] = 3
```

```python
train_df.loc[train_df['Fare'] < 30, 'Fare'] = 0
train_df.loc[(30<=train_df['Fare']) & (train_df['Fare']<80), 'Fare'] = 1
train_df.loc[(80<=train_df['Fare']) & (train_df['Fare']<100), 'Fare'] = 2
train_df.loc[(100<=train_df['Fare']), 'Fare'] = 3

test_df.loc[test_df['Fare'] < 30, 'Fare'] = 0
test_df.loc[(30<=test_df['Fare']) & (test_df['Fare']<80), 'Fare'] = 1
test_df.loc[(80<=test_df['Fare']) & (test_df['Fare']<100), 'Fare'] = 2
test_df.loc[(100<=test_df['Fare']), 'Fare'] = 3
```

```python
train_df.drop(['PassengerId', 'Name', 'Ticket'], axis=1, inplace=True)
test_df.drop(['Name', 'Ticket'], axis=1, inplace=True)
display(train_df.head(3))
display(test_df.head(3))
```

```python
x_data = train_df.drop(['Survived'], axis=1)
t_data = train_df['Survived'].values.reshape(-1, 1)

X = tf.placeholder(shape=[None, 7], dtype=tf.float32)
T = tf.placeholder(shape=[None, 1], dtype=tf.float32)

W = tf.Variable(tf.random.normal([7,1], name='weight'))
b = tf.Variable(tf.random.normal([1]), name='bias')

logit = tf.matmul(X,W) + b
H = tf.sigmoid(logit)

loss = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(logits=logit,
                                                             labels=T))

train = tf.train.GradientDescentOptimizer(learning_rate=1e-4).minimize(loss)

sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(350000):
    _, loss_val = sess.run([train, loss], feed_dict={X:x_data,
                                                    T:t_data})
    if step % 35000==0:
        print('loss : {}'.format(loss_val))
```

```python
temp = test_df.drop(['PassengerId'], axis='columns')

result_dict={'PassengerId':[], 'result':[]}

for i in range(len(temp)):
    predict_data = [list(temp.iloc[i])]
    
    result = sess.run(H, feed_dict={X:predict_data})
    result_dict['PassengerId'].append(test_df.PassengerId.iloc[i])
    result_dict['result'].append(result[0][0])

for i in range(len(result_dict['result'])) :
    if result_dict['result'][i]<0.5:
        result_dict['result'][i]=0
    else:
        result_dict['result'][i]=1


final_df = pd.DataFrame(result_dict)

final_df.rename(columns={'result':'Survived'}, inplace=True)
display(final_df)
```

```python
final_df.to_csv('submission.csv', sep=',', na_rep='NaN', index=False)
```
