```python
# 2. 각 사용자별 평균 평점 (출력시 정렬은 userId로 오름차순 정렬)
import pandas as pd

df = pd.read_csv(r'C:\Users\sun\Desktop\Data\movie\ratings.csv')
group_user = df.groupby(['userId']).mean()
display(group_user)
print(group_user['rating'])
```

```python
# 3. 각 영화별 평균 평점 (출력시 정렬은 movieId로 오름차순 정렬)
import pandas as pd

df = pd.read_csv(r'C:\Users\sun\Desktop\Data\movie\ratings.csv')
group_movie = df.groupby(['movieId']).mean()
display(group_movie)
print(group_movie['rating'])
```

```python
# 4. 평균 평점이 가장 높은 영화의 제목
# (단, 동률이 있을 경우 모두 출력하고 title을 기준으로 오름차순 정렬)
import pandas as pd

movie_df = pd.read_csv(r'C:\Users\sun\Desktop\Data\movie\movies.csv')
rating_df = pd.read_csv(r'C:\Users\sun\Desktop\Data\movie\ratings.csv')

group_id = rating_df.groupby(['movieId']).mean()
movie_df.set_index('movieId', inplace=True)
df = pd.merge(movie_df, group_id, on='movieId', how='inner')
# display(df)

maximum = df['rating'].max()
result = df.loc[df['rating']==maximum]
display(result)

sort_result = result.sort_values(by='title')
print(sort_result['title'])
```

```python
# 5. Comedy영화 중 가장 평점이 낮은 영화의 제목을 출력
# (단, 동률이 있을 경우 모두 출력하고 title을 기준으로 오름차순 정렬)
import pandas as pd

movie_df = pd.read_csv(r'C:\Users\sun\Desktop\Data\movie\movies.csv')
rating_df = pd.read_csv(r'C:\Users\sun\Desktop\Data\movie\ratings.csv')

group_id = rating_df.groupby(['movieId']).mean()
movie_df.set_index('movieId', inplace=True)
df = pd.merge(group_id, movie_df, on='movieId', how='inner')
# display(df)

cond1 = df['genres'].str.contains('Comedy')
comedy_df = df.loc[cond1, :]
# display(comedy_df)

minimum = comedy_df['rating'].min()
result = comedy_df.loc[comedy_df['rating']==minimum]
sort_result = result.sort_values(by='title')
display(sort_result)
print(sort_result['title'])
```

```python
# 6. 2015년도에 평가된 모든 Romance 영화의 평균 평점
import pandas as pd

movie_df = pd.read_csv(r'C:\Users\sun\Desktop\Data\movie\movies.csv')
rating_df = pd.read_csv(r'C:\Users\sun\Desktop\Data\movie\ratings.csv')

# group_id = rating_df.groupby(['movieId']).mean()    # 영화별 평균 먼저하면 안됨
# 로맨스 영화 따로 빼내고 거기서 평균

movie_df.set_index('movieId', inplace=True)
df = pd.merge(movie_df, rating_df, on='movieId', how='inner')
# display(df)

df['timestamp'] = pd.to_datetime(df['timestamp'], unit='s')
df['year'] = df['timestamp'].dt.year
# display(df)

year_df = df.loc[df['year']==2015]
display(year_df)

result = year_df.loc[year_df['genres'].str.contains('Romance'), :]
display(result)
print(result['rating'].mean())
```

```python
# 7. 모든 영화장르 중 사용자 평점이 가장 좋은 영화장르
# (단, 동률이 있을 경우 장르를 기준으로 오름차순 정렬)
import pandas as pd

movie_df = pd.read_csv(r'C:\Users\sun\Desktop\Data\movie\movies.csv')
rating_df = pd.read_csv(r'C:\Users\sun\Desktop\Data\movie\ratings.csv')

movie_df.set_index('movieId', inplace=True)
df = pd.merge(movie_df, rating_df, on='movieId', how='inner')
# display(df)

group_gen = df.groupby(['genres']).max()
display(group_gen)

maximum = group_gen['rating'].max()
result = group_gen.loc[group_gen['rating']==maximum]
print(result['rating'])
```
